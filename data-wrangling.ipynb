### Column removals

##### Columns 'Unnamed: 22'  and 'Unnamed: 23'  are unclear to me. What kind of data is in there? 
##### Let's have a look and see if we can optimize them. 

```python
# Let's check how many NULL values they have, of a total of 5992 rows.

sharks_copy.isna().sum()/len(sharks_copy)

# Unnamed: 22 and Unnamed: 23 columns have almost 100% of NULL values. I have decided to remove them from the 
# dataframe, because I don't see the value of keeping the columns
```

```python
# I want to remove columns Unnamed: 22 and Unnamed: 23

sharks_copy = sharks_copy.drop(columns=['unnamed:_22', 'unnamed:_23'])
```

### Column: 'sex'

##### The 'sex' column seems to have information we can work with, and potentially improve

```python
# how many unique values does the 'sex' column have?

sharks_copy["sex"].nunique() 
```

```python
# how often do they occur and can we optimize these values?

sharks_copy.describe(include=[object])

sharks_copy['sex'].value_counts()
```

```python
# seems that there is an unnecessary space in 2 of the values, let's remove the space and add them to the 'M' values

sharks_copy['sex'] = sharks_copy['sex'].str.replace('M ', 'M')
```

```python
sharks_copy['sex'].value_counts()
```

```python
# let's look into the strange values and see if we can replace them with the right value or make them 'Unknown'

# Name = Brian Kang, we can assume this is a guy's name and will replace it with an 'M'
sharks_copy.loc[sharks_copy['sex'] == 'lli'] 
sharks_copy['sex'] = sharks_copy['sex'].str.replace('lli', 'M')

# unknown name, can't deduct if it's a female/male/neutral. Therefore changing into Unknown
sharks_copy.loc[sharks_copy['sex'] == '.'] 
sharks_copy['sex'] = sharks_copy['sex'].str.replace('.', 'Unknown')

# 2 people on a boat got attacked: William & Leslie Newton. This could be a maried couple (man+female) 
# or 2 brothers (2x male). For me the gender is unclear and thus want to change the N to Unknown
sharks_copy.loc[sharks_copy['sex'] == 'N'] 
sharks_copy['sex'] = sharks_copy['sex'].str.replace('N', 'Unknown')
```

```python
sharks_copy['sex'].value_counts()
```

##### Compare the 'sex' and 'name' column and see if we find information that can help to improve the data

```python
# checking the name column to see if we could deduct if the attacked persons were male or female and replace
# the NULL values of the 'sex' column with better information deducted from the 'name' column

sharks_copy['name'].value_counts()
```

```python
sharks_copy.loc[sharks_copy['name'] == 'male']
```

```python
# match between 'male' in 'name' and a Null value in 'sex': 9x. 
# Therefore we can replace the missing value in 'sex' with male
# Won't touch the 'name' column further for now

sharks_copy.loc[(sharks_copy['name'] == 'male') & (sharks_copy['sex'].isna()), 'sex'] ='M'

# sharks_copy.loc[(sharks_copy['name'] == 'female') & (sharks_copy['sex'].isnull()==True)] # nothing
# sharks_copy.loc[(sharks_copy['name'] == 'boy') & (sharks_copy['sex'].isnull()==True)] # nothing
```

```python
sharks_copy['sex'].value_counts()
```

### Column: 'year'

##### Check the data and and see if we could categorize them

```python
sharks_copy.describe()
```

```python
# let's see what values are in the column

sharks_copy['year'].value_counts()
```

```python
print(set(sharks_copy['year']))

# sporadically checked parts of the data and years without 4 digits seem to be actual years. 
# Therefore, for now, assuming they could be actual years and won't change them
```

```python
# categorize years 'before-1900', '1901 and 1950', '1951 and 2000', '2001-after'
# create new column called 'year_category'


sharks_copy = sharks_copy.assign(year_category=np.where(sharks_copy["year"].le(1901), 
                                               "before-1900", 
                                               np.where(sharks_copy["year"].le(1951), 
                                                        "1901-1950", 
                                                        np.where(sharks_copy["year"].le(2001),
                                                                 "1951-2000",
                                                                    "2001-after"))))
```
